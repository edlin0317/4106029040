### load datasets

```python
from sklearn import datasets
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
```

```python
plt.scatter(iris.data[:,:1],iris.data[:,1:2],c=iris.target, cmap=plt.cm.Dark2)
plt.title('Sepal plot')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
plt.scatter(iris.data[:,2:3],iris.data[:,3:4],c=iris.target, cmap=plt.cm.Dark2)
plt.title('Sepal plot')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

### test & train split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['target']), df['target'], test_size=0.33, random_state=87)
```

### train & predict & measure

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
pred = model.predict(X_test)
print (accuracy_score(y_test, pred))
```
